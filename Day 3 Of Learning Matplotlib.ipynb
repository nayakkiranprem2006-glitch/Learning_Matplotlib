<b>Questions:
1. Plot a line graph with red dashed lines and circle markers.
2. Change x and y axis limits manually.
3. Add grid lines to a plot.
4. Customize tick labels and font sizes.</b>

---

```python
# Q1. Plot a line graph with red dashed lines and circle markers.

import matplotlib.pyplot as plt

classes = [1,2,3,4,5]
score_A = [20,30,40,50,60]


plt.plot(classes, score_A, color="Red", linestyle = "--", marker = "o")
plt.xlabel("Classes")
plt.ylabel("Score")
plt.title("Score of the class")
plt.grid(True)
plt.legend(["Sales 2025"])
plt.show()
```

```python
# Q2. Change x and y axis limits manually.

import matplotlib.pyplot as plt

classes = [1,2,3,4,5]
score_A = [20,30,40,50,60]


plt.plot(classes, score_A, color="Red", linestyle = "--", marker = "o")
plt.xlabel("Classes")
plt.ylabel("Score")
plt.xlim(0.5,5.5)
plt.ylim(5,100)
plt.title("Score of the class")
plt.grid(True)
plt.legend(["Sales 2025"])
plt.show()
```

```python
# Q3. Add grid lines to a plot.

import matplotlib.pyplot as plt

classes = [1,2,3,4,5]
score_A = [60,50,40,30,20]


plt.plot(classes, score_A, color="Red", linestyle = "--", marker = "o")
plt.xlabel("Classes")
plt.ylabel("Score")
plt.title("Score of the class")
plt.grid(True)
plt.legend(["Sales 2025"])
plt.show()
```

```python
# Q4. Customize tick labels and font sizes.
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 13})

classes = [1,2,3,4,5]
score_A = [20,30,40,50,60]


plt.plot(classes, score_A, color="Red", linestyle = "--", marker = "o")
plt.xlabel("Classes")
plt.ylabel("Score")
plt.xticks([1,2,3,4,5], ["One", "Two", "Three", "Four", "Five"])
plt.yticks([20,30,40,50,60], ["TWENTY", "THIRTY", "FORTY", "FIFTY", "SIXTY"])
plt.title("Score of the class")
plt.grid(True)
plt.legend(["Sales 2025"])
plt.show()
```
