```python
# Reminder what should we have learn in our recent class

import matplotlib.pyplot as plt

x= [1,2,3,4,5]
y= [10,52,34,96,88]

plt.plot(x,y, linestyle= "-.", marker= "o")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Sales Charts")
plt.legend("Sales", loc="upper left")
plt.show()
```

```python
# Q1. Create a bar chart of students and their marks.

import matplotlib.pyplot as plt
student =['Aman', 'Babli', 'Chintu', 'Daksh']
marks = [70, 65, 84, 49]

plt.bar(student, marks, color= "orange")
plt.legend(['Result 2025'], loc="upper right")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title('Result Comparison')
plt.grid(color= 'Black', linestyle= "--", linewidth = 0.5)
plt.show()
```

```python
# Q2. Create a horizontal bar chart for product sales.

import matplotlib.pyplot as plt
student =['Aman', 'Babli', 'Chintu', 'Daksh']
marks = [70, 65, 84, 49]

plt.barh(student, marks, color= "orange")
plt.legend(['Result 2025'], loc="upper right")
plt.ylabel("Student")
plt.xlabel("Marks")
plt.title('Result Comparison')
plt.grid(color= 'gray', linestyle= "--", linewidth = 0.5)
plt.show()
```

```python
# Q3. Create a pie chart showing market share of 4 companies.

import matplotlib.pyplot as plt
company= ["Company A", "Company B", "Company C", "Company D"]
share= [2400, 4500, 3500, 1900]
colours= ['Red', 'Aqua', "Green", "Blue"]

plt.pie(share, labels= company, autopct= "%1.1f%%", colors = colours)
plt.title("Market Share of Campanies")
plt.axis('equal')
plt.show()
```

```python
# Q4. Plot a histogram of ages: [22,25,30,22,35,40,28,25,30].

import matplotlib.pyplot as plt

ages = [22,25,30,22,35,40,28,25,30]

plt.hist(ages, bins=5, color="violet", edgecolor= 'black')
plt.xlabel("Age of Students")
plt.ylabel("No. of Students")
plt.title("Age Distribution of Student")
plt.show()
```

```python
# Q5. Plot a scatter plot of height vs weight.

import matplotlib.pyplot as plt

weight=[45,50,55,60,65,42,68]
age= [20,24,26,30,40,45,60]

plt.scatter(weight,age, linestyle= "--", color= "Red", marker="o", label="Weight vs Age")

plt.xlabel("Weight")
plt.ylabel("Age")
plt.legend()
plt.title("Comparison b/w Weight and Age")
plt.grid(True)
plt.show()
```
