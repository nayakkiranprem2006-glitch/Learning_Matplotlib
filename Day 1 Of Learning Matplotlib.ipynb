Questions:
1. Plot a line graph for x = [1,2,3,4], y = [10, 20, 30, 40].
2. Add a title "Sales Over Time" and label the axes.
3. Add a legend called "Sales".
4. Save the plot as "sales_plot.png".


```python
#  Q1: Line Plot
# Plot a line graph for:
# python
# x = [1, 2, 3, 4]
# y = [10, 20, 30, 40]

# Plot the graph using plt.plot(x, y) and display it.

import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.show()
```

```python
# Q2: Add labels and a title to the plot.

# Use the same x and y data:
# - X-axis label: "X Axis"
# - Y-axis label: "Y Axis"
# - Title: "Sales Over Time"

import matplotlib.pyplot as plt

x= [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.title("Sales Over Time")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# Q3. Add a legend called "Sales".

import matplotlib.pyplot as plt

x= [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.title("Sales Over Time")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(["Sales"], loc="upper left")
plt.show()
```

```python
# 4. Save the plot as "sales_plot.png".
import matplotlib.pyplot as plt

x= [1,2,3,4]
y = [10,20,30,40]

plt.plot(x,y)
plt.title("Sales Over Time")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(["Sales"], loc="upper left")
plt.savefig("Sales_Over_Time.png")
plt.show()
```
