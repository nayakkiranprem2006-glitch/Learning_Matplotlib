```python
# Q1. Plot 2 line plots side-by-side using subplot().

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,30,15,25,45]

plt.subplot(1,2,1) #plt.subplot(1st Row, 2nd Column, 1st Subplot)
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) #plt.subplot(1st Row, 2nd Column, 2nd Subplot)
plt.bar(x,y)
plt.title('Bar Chart')

plt.show()
```

```python
# Q2. Plot 4 different charts in a 2x2 grid.

import matplotlib.pyplot as plt
x = ["Comp1","Comp2","Comp3","Comp4","Comp5"]

y = [10,30,15,25,45]

plt.subplot(2, 2, 1)
plt.plot(x,y)

plt.subplot(2, 2, 2)
plt.bar(x, y, color= "Purple")

plt.subplot(2, 2, 3)
plt.scatter(x, y, marker="^", linestyle= "-.", color="red")

plt.subplot(2, 2, 4)
plt.pie(y, labels=x, autopct= "%1.1f%%")

plt.tight_layout()
plt.show()
```

```python
# Q3. Create a figure of custom size (8,6).

plt.figure(figsize= (8, 6))
x = ["Comp1","Comp2","Comp3","Comp4","Comp5"]

y = [10,30,15,25,45]
plt.plot(x, y, marker="^", color= "red", linestyle= "-.")
plt.grid(True)
plt.show()
```

```python
# Q4. Use tight_layout() to fix overlapping.
import matplotlib.pyplot as plt
x = ["Comp1","Comp2","Comp3","Comp4","Comp5"]

y = [10,30,45,70,80]

plt.subplot(2, 2, 1)
plt.plot(x,y)

plt.subplot(2, 2, 2)
plt.bar(x, y, color= "Purple")

plt.subplot(2, 2, 3)
plt.scatter(x, y, marker="^", linestyle= "-.", color="red")

plt.subplot(2, 2, 4)
plt.pie(y, labels=x, autopct= "%1.1f%%")

plt.tight_layout()
plt.show()
```
